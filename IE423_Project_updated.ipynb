# Aim

The project aims to use statistical quality control techniques, such as CUSUM and Shewhart, to monitor and analyze soccer match data. The primary objectives are twofold: to detect shifts in key metrics resulting from specific game events and to evaluate the timeliness of these detections.

In this project, the analysis begins by identifying "out-of-control" statistics using control charts. These detected anomalies are then linked to specific events that occurred earlier in the game, establishing cause-and-effect relationships. In Task 2, the focus shifts to examining these "out-of-control" statistics to uncover potential patterns or relationships that may predict future events in the game. These predicted relationships are then validated by observing their occurrences in the actual gameplay.

Ultimately, the aim is to bridge the principles of quality engineering within the soccer analytics context. Through descriptive and preventive analyses, the aim is to provide a systematic approach to interpreting and predicting significant game developments.

# Methodology

"Number of Attacks" (Attacks_home) is selected as the primary metric for analysis. We only executed this metric on the home team as the events caused by the away team can also lead to out of control statistics for Attacks_home. This choice was motivated by the ability of this metric to facilitate the construction of cause-and-effect relationships. Also, identifying "out-of-control" attacks provides a clearer opportunity to link these anomalies to significant upcoming events in the game, making the analysis more insightful.

For Task 1, it was considered that the number of attacks might go out of control in the following scenarios:  
- When the away team scored an equalizing or leading goal.  
- When the away team received a red card, increasing the pressure exerted by the home team. 
- When the away team scored subsequent goals which would lead to an increase in the pressure of the home team

For Task 2, it was hypothesized that out-of-control attacks could result in the following outcomes:  
- A goal being scored by the home team
- A penalty being awarded to the home team

To obtain a statistical control chart, a CUSUM chart was utilized for the following reasons: CUSUM provides a cumulative measure of increases in the number of attacks. Therefore, when the chart exceeds the decision boundary, it indicates that the home team is executing a series of sustained attacks, supporting the hypotheses outlined above. While other control charts, such as the Shewhart chart, could also be used, they primarily focus on significant increases in attacks at a specific moment. However, they may not effectively capture the pressure of a team, which reflects sustained increases in attacks over a timeframe, nor are they as suitable for detecting small mean shifts in the context of quality control.

To construct the CUSUM chart, the tabular CUSUM approach was employed, and the chart was developed as follows:
- The Attacks_home values were given as cumulative, which are converted to attacks that occured each minute as "Attacks_home_each_min". 
- The target value (μ_0) was calculated as the average of the "Attacks" feature for each match individually.
- The value of mu_1 was determined as μ_0 + 1 * σ where 1 * σ was deemed appropriate for this particular analysis.
- The reference value (K) was calculated using the formula: K = |μ_1 - μ_0| / 2.
- The values of C_positive_i and C_negative_i (accumulated deviations from μ_0) were computed. These quantities reset to zero whenever their values became negative. While C_negative_i values were also calculated, they were not considered insightful since μ_0 is close to zero, and the "Attacks" metric cannot go below zero. The formulas used were:
    - C_positive_i = max[0, x_i - (μ_0 + K) + C_plus_i_-1]
    - C_negative_i = max[0, (μ_0 - K) + C_negative_i_-1]
- Finally, a decision interval H = 4 * σ is decided upon. If the C_positive_i value goes beyond that threshold , it indicates a mean shift at that point.

After constructing the CUSUM charts for each match, key events are plotted as vertical lines on the graphs to aid in interpreting the causes of mean shifts. Specifically, goals scored by both the home and away teams, red cards issued to the away team, and penalties awarded to the home team are included in the plots. These visualizations are essential for executing Task 1 and Task 2, as they help establish connections between statistical shifts and significant game events.


# Code

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("matches.csv")
df
```

```python
df['Attacks_home_each_min'] = df.groupby('fixture_id')['Attacks_home'].diff()
df['Attacks_home_each_min'] = df['Attacks_home_each_min'].fillna(df['Attacks_home']).astype(int)

df["Attacks_away_each_min"] = df.groupby('fixture_id')['Attacks_away'].diff()
df['Attacks_away_each_min'] = df['Attacks_away_each_min'].fillna(df['Attacks_away']).astype(int)
```

```python
grouped_data = df.groupby('fixture_id')['Attacks_home_each_min']

# Calculate mu_0 and sigma for each match
match_stats = grouped_data.agg(mu_0='mean', sigma='std').reset_index()

# Calculate mu_1 and K for each match
match_stats['mu_1'] = match_stats['mu_0'] + match_stats['sigma']
match_stats['K'] = (match_stats['mu_1'] - match_stats['mu_0']) / 2

# Initialize C_plus and C_minus columns
df['C_plus'] = 0
df['C_minus'] = 0

for fixture_id, group in df.groupby('fixture_id'):

    mu_0 = match_stats.loc[match_stats['fixture_id'] == fixture_id, 'mu_0'].values[0]
    K = match_stats.loc[match_stats['fixture_id'] == fixture_id, 'K'].values[0]
    
    C_plus = 0
    C_minus = 0
    
    for idx, row in group.iterrows():
        x_i = row['Attacks_home_each_min']
        
        # Update C_plus and C_minus based on the formulas
        C_plus = max(0, x_i - (mu_0 + K) + C_plus)
        C_minus = -max(0, (mu_0 - K) - x_i + C_minus)
        
        df.loc[idx, 'C_plus'] = C_plus
        df.loc[idx, 'C_minus'] = C_minus
        
# Decision interval H
match_stats['H'] = 4 * match_stats['sigma']  
```

```python
def plot_cusum(data):
    for fixture_id, group in data.groupby('fixture_id'):
        # Get the corresponding H value
        H = match_stats.loc[match_stats['fixture_id'] == fixture_id, 'H'].values[0]
        
        plt.figure(figsize=(10, 6))
        plt.plot(group['minute'], group['C_plus'], label='C_plus', color='blue')
        plt.plot(group['minute'], group['C_minus'], label='C_minus', color='red')
        plt.axhline(y=H, color='green', linestyle='--', label='H (Decision Interval)')
        
        away_goal_mins = group[group['Goals_away'].diff() > 0]['minute']    
        for minute in away_goal_mins:
            plt.axvline(x=minute, color='purple', linestyle='-', label='Away Team Goal')
           
        home_goal_mins = group[group['Goals_home'].diff() > 0]['minute']    
        for minute in home_goal_mins:
            plt.axvline(x=minute, color='green', linestyle='-', label='Home Team Goal')
            
        red_card_minutes = group[group['Redcards_away'].diff() > 0]['minute']
        for minute in red_card_minutes:
            plt.axvline(x=minute, color='red', linestyle='-', label='Away Team Red Card')
            
        penalty_mins = group[group['Penalties_home'].diff() > 0]['minute']
        for minute in penalty_mins:
            plt.axvline(x=minute, color='black', linestyle='-', label='Home Team Wins Penalty')
            
        plt.title(f'CUSUM Chart for Match {fixture_id}')
        plt.xlabel('Minute')
        plt.ylabel('CUSUM Value')
        plt.legend()
        plt.grid()
        plt.show()
```

```python
# At this point, all 102 mathces are examined. However 6 significant matches are selected for further analysis.
selected_fixture_ids = [19135254, 19155142, 19155162, 19155178,
                        19172022, 19172038]

filtered_data = df[df['fixture_id'].isin(selected_fixture_ids)]
plot_cusum(filtered_data)
```

The implementation of CUSUM (Cumulative Sum Control Charts) in this project aligns closely with the lecture materials on Statistical Process Control (SPC) and time-series data analysis. The lectures emphasized the importance of detecting small shifts in processes, which traditional methods like Shewhart charts might fail to capture. CUSUM was highlighted as a superior method, especially in scenarios where data exhibit serial correlation. This project effectively demonstrates these concepts by applying CUSUM to monitor attack patterns in football matches, providing a real-world example of its utility.

Calculating control limits (UCL and LCL) and defining a central line were fundamental to the project's methodology, reflecting the theoretical foundations of SPC discussed in the lectures. These control limits are critical for identifying significant deviations from the norm, enabling a structured and systematic analysis of process stability. By including these elements, the project reinforces the role of SPC in ensuring robust monitoring and timely detection of changes in process performance.

The project also addresses autocorrelation in data, a common issue in process analysis that can impact the reliability of traditional control charts. The lecture materials emphasized that manufacturing and dynamic systems often exhibit inherent correlations, which must be accounted for to avoid misleading results. By using cumulative sums, the project accommodates these dependencies, ensuring that insights derived from the analysis are both robust and reliable.

CUSUM itself is a powerful tool designed to monitor small, consistent shifts in a process by accumulating deviations over time. Unlike traditional methods that provide static snapshots, CUSUM offers a dynamic perspective, making it particularly suitable for analyzing continuous or time-dependent data such as football attack rates. In this project, CUSUM demonstrated its potential to detect momentum changes aligned with critical match events like goals scored or red cards. This provided clear visualizations of attack intensity trends, enabling actionable insights for teams and analysts while improving the understanding of how external factors influence performance metrics.

Reflecting on the results, CUSUM proved highly effective in highlighting key moments where attack patterns exceeded control thresholds, revealing strategic opportunities or vulnerabilities.
```python
def plot_attacks_per_min_home(data):

    grouped = data.groupby('fixture_id')
    
    for fixture_id, group in grouped:
        
        # CL, UCL, LCL for home team
        CL_home = group['Attacks_home_each_min'].mean()
        UCL_home = CL_home + 3 * CL_home ** 0.5
        LCL_home = CL_home - 3 * CL_home ** 0.5
        

        plt.figure(figsize=(10, 6))
        plt.plot(group['minute'], group['Attacks_home_each_min'], color="blue", marker='o', label='Number of Attacks (home)')
        
        plt.axhline(y=CL_home, color='green', linestyle='--', label="Centerline (home)")
        plt.axhline(y=UCL_home, color='red', linestyle='--', label="UCL (home)")
        plt.axhline(y=max(0,LCL_home), color='red', linestyle='--', label="LCL (home)")
        
        away_goal_mins = group[group['Goals_away'].diff() > 0]['minute']    
        for minute in away_goal_mins:
            plt.axvline(x=minute, color='purple', linestyle='-', label='Away Team Goal')
           
        home_goal_mins = group[group['Goals_home'].diff() > 0]['minute']    
        for minute in home_goal_mins:
            plt.axvline(x=minute, color='green', linestyle='-', label='Home Team Goal')
            
        red_card_minutes = group[group['Redcards_away'].diff() > 0]['minute']
        for minute in red_card_minutes:
            plt.axvline(x=minute, color='red', linestyle='-', label='Away Team Red Card')
        
        plt.title(f"Attacks Per Minute for Fixture ID: {fixture_id}", fontsize=14)
        plt.xlabel("Minute", fontsize=12)
        plt.ylabel("Attacks (Home) Per Minute", fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()
        
        # Show the plot
        plt.show()
```
